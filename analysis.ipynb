```python
## 1. Import libraries and load data, ت
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## 2. Data exploration and cleaning
df = pd.read_csv('ecommerce_customer_churn_dataset.csv')
print(df)
print(df.head())
print(df.shape)
print(df.info())
df.describe()
df['Credit_Balance'].fillna(df['Credit_Balance'].mean(), inplace=True)
print(df.isnull().sum())
df.drop_duplicates(inplace=True)
print(df.shape)

## 3. Customer segmentation
def classify_customer(freq):
    if freq >= 20:
        return 'Active'
    elif freq >= 10:
        return 'Average active'
    else:
        return 'Sleeping'

## 4. Exploratory data analsis (EDA)    
df['Customer_Segment'] = df['Login_Frequency'].apply(classify_customer)  
print(df['Customer_Segment'] .value_counts())  
import seaborn as sns
plt.figure(figsize=(8, 5))
df['Customer_Segment'].value_counts().plot(kind='bar', color=['green', 'orange', 'red'])
plt.title('Customer classification')
plt.xlabel('Category')
plt.ylabel('Number')
plt.tight_layout()
plt.show()
plt.Figure(figsize=(12, 6))
df.groupby('Country')['Churned'].mean().sort_values(ascending=False).plot(kind='bar', color='red')
plt.title('Percentage of customers quitting by country')
plt.xlabel('Country')
plt.ylabel('Percentage of leaving')
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
plt.scatter(df['Session_Duration_Avg'], df['Cart_Abandonment_Rate'], alpha=0.3, color='blue')
plt.title('The correlation between session duration and cart adandonment rate')
plt.xlabel('Average session duration')
plt.ylabel('Cart adandonment rate')
plt.tight_layout()
plt.show()
plt.figure(figsize=(8, 5))
df['Age'].hist(bins=20, color='green', edgecolor='black')
plt.title('Customer distribution by age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

## 5. AI powered recommendations
import os
from groq import Groq
from dotenv import load_dotenv


load_dotenv()
secret_api_key = os.getenv("MY_KEY")

client = Groq(api_key=secret_api_key)

segments = df['Customer_Segment'].value_counts().to_dict()

for segment, count in segments.items():
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{
            "role": "user",
            "content": f"You are an e-commerce data analyst. We have {count} customers in the '{segment}' segment. Write a concise marketing recommendation in 3 sentences for this segment."
        }]
    )
    print(f"\n--- Recommendation for Segment: {segment} ---")
    print(response.choices[0].message.content)
```
